```python
#imports
import cadquery as cq
from cadquery.vis import show
```

```python
flame_outer=cq.importers.importDXF("assets/svg/dxf_derivatives/flame_outer.dxf")
flame_outer_3d = cq.Workplane("XY").add(flame_outer).extrude(3.0)
```

```python
show(flame_outer_3d)
```

```python
def bounding_box_dimensions(name,bounding_box):
    print(f"{name}'s bounding box: ({bounding_box.xlen},{bounding_box.ylen},{bounding_box.zlen})")
```

```python
flame_inner = cq.importers.importDXF("assets/svg/dxf_derivatives/flame_inner.dxf")
flame_inner_3d = cq.Workplane("XY").add(flame_inner).extrude(2.0)
```

```python
show(flame_inner_3d)
```

```python
flame_full_3d = flame_outer_3d.union(flame_inner_3d.translate((7,0,3.0)))
```

```python
show(flame_full_3d)
```

```python
bicep_mirror_filled = cq.importers.importDXF("assets/svg/dxf_derivatives/bicep_mirror_filled.dxf")
bicep_mirror_filled_3d = cq.Workplane("XY").add(bicep_mirror_filled).extrude(3).val().scale(0.15)
```

```python
bicep_mirror_filled_3d.BoundingBox().xlen
```

```python
show(bicep_mirror_filled_3d)
```

```python
flame_with_bicep = flame_full_3d.union(bicep_mirror_filled_3d.translate((12,2,5.0)))
```

```python
show(flame_with_bicep)
```

```python
bounding_box_dimensions("Flame_outer",flame_outer_3d.val().BoundingBox())
bounding_box_dimensions("Flame_inner",flame_inner_3d.val().BoundingBox())
bounding_box_dimensions("Bicep",bicep_mirror_filled_3d.BoundingBox())
```

```python
# flame_with_bicep_keychain = flame_with_bicep.faces(">Z").workplane().moveTo(14,25).hole(4.0)
textA = cq.Workplane().text("A",fontsize=4,combine=True,distance=0.5012,fontPath="assets/fonts/GreatVibes-Regular.ttf")

text_implies = cq.Workplane().text("=>",fontsize=6,combine=True,distance=0.5012,fontPath="assets/fonts/GreatVibes-Regular.ttf")
```

```python
flame_with_bicep_and_letters = flame_with_bicep.union(textA.translate((13,12,5.0))).union(text_implies.translate((14,8,5.0)))
```

```python
show(flame_with_bicep_and_letters)
```

```python
flame_with_bicep_keychain = flame_with_bicep_and_letters.faces(">Z").workplane().moveTo(14,25).hole(4.0)
```

```python
show(flame_with_bicep_keychain)
```
